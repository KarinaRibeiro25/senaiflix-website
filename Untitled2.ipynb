```python
import random
numero_ia = random.randint(1,5)
print("--- JOGO DA IA ---") #informativo
print("Eu pensei em um número de 1 a 5. Você tem três chances para adivinhar")

contador = 0

while True:
    if contador < 3:
        palpite = int(input("Qual é o seu palpite?"))
        if palpite == numero_ia:
            print("UAU! Você acertou!")
            break #Quebra o laço de repetição se acertar

        else:
            print("Errou! Tente novamente.")
            break

    else:
        print("Excedeu as tentativas.")
        break
    contador += 1
    print(contador)

print(f"O número secreta era {numero_ia}.")
```
