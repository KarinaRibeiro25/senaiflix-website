```python
# --- Tomando decisões (if/else) ---
humor = input("Como você está se sentindo hoje? (feliz/triste/indiferente) ")
if humor == "feliz":
    print("Que ótimo! Computadores adoram humanos felizes!")
elif humor == "indiferente": # para colocar mais de uma condição
    print("Tudo bem então, nem feliz nem triste")
else:
    print("Poxa, espero que aprender Python melhore o seu dia!")
```

```python
# --- laços de Repetições ( for e while) ---
print("Iniciando a varredura do sistema...")
# O laço "for" vai repetir o código recuado 3 vezes
for numero in range(3):
    print("Analisando número", numero)

print("Varredura concluída!")
```

```python
# --- laços de Repetições ( for e while) ---
print("Iniciando a varredura do sistema...")
# O laço "for" vai repetir o código recuado 3 vezes
for numero in range(1,10):# delimitar a quantidade de itens na lista por uma quantidade especifica
    print("Analisando número", numero)

print("Varredura concluída!")
```

```python
# --- laços de Repetições ( for e while) ---
print("Iniciando a varredura do sistema...")
# O laço "for" vai repetir o código recuado 3 vezes
for numero in range(0,101,2): # delimitar e alternar numero na lista
    print("Analisando número", numero)

print("Varredura concluída!")
```
